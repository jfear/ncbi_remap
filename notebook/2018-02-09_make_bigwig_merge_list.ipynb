# BigWig Merge List

I want to go ahead and merge together what files I have from the FlyBase list.

```python
from pathlib import Path
import pandas as pd
from itertools import zip_longest
```

```python
fb = pd.read_csv('../output/flybase_samples.tsv', sep='\t').srx.tolist()
```

```python
all_first = []
all_second = []
for srx in fb:
    first = Path(f'~/Projects/ncbi_remap/output/aln-wf/samples/{srx}/{srx}.first.bw').expanduser()
    second = Path(f'~/Projects/ncbi_remap/output/aln-wf/samples/{srx}/{srx}.second.bw').expanduser()
    
    if first.exists() & second.exists():
        all_first.append(str(first))
        all_second.append(str(second))
    else:
        print(srx)
```

```python
def grouper(iterable, n, fillvalue=''):
    "Collect data into fixed-length chunks or blocks"
    # grouper('ABCDEFG', 3, 'x') --> ABC DEF Gxx"
    args = [iter(iterable)] * n
    return list(zip_longest(*args, fillvalue=fillvalue))
```

```python
Path('../bin/run_flybase_bigwig_merge').mkdir(exist_ok=True)
```

```python
bnum = 50
```

```python
len(grouper(all_first, bnum))
```

```python
for i, block in enumerate(grouper(all_first, bnum)):
    with open(f'../bin/run_flybase_bigwig_merge/flybase_first_strand_list_block{i}.sh', 'w') as fh:
        fh.write('#!/bin/bash\n')
        inputs = ' '.join(block)
        output = f'/scratch/fearjm/first_block{i}.bedGraph'
        cmd = f'bigWigMerge {inputs} {output}\n'
        fh.write(cmd)
```

```python
'bigWigMerge ' + ' '.join([f'/scratch/fearjm/first_block{i}.bedGraph' for i in range(bnum)])
```

```python
for i, block in enumerate(grouper(all_first, bnum)):
    with open(f'../bin/run_flybase_bigwig_merge/flybase_second_strand_list_block{i}.sh', 'w') as fh:
        fh.write('#!/bin/bash\n')
        inputs = ' '.join(block)
        output = f'/scratch/fearjm/second_block{i}.bedGraph'
        cmd = f'bigWigMerge {inputs} {output}\n'
        fh.write(cmd)
```
