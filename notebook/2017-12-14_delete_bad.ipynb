# Delete bad files

I am really tight on disk space, so I am going to go through here and delete FASTQ and BAM files from "bad" samples.

```python
import os
import sys
from pathlib import Path

import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Project level imports
sys.path.insert(0, '../lib')
from ncbi_remap.notebook import Nb
from ncbi_remap.plotting import make_figs

# Setup notebook
nbconfig = Nb.setup_notebook()

# Turn on cache
from joblib import Memory
memory = Memory(cachedir=nbconfig.cache, verbose=0)

# Connect to data store
store = pd.HDFStore('../output/sra.h5', mode='r')
```

```python
def clean(srx=None, srr=None):
    def _delete(path):
        if Path(path).exists():
            print('Removing: {}'.format(path))
            Path(path).unlink()
            
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}_1.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}_2.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}_1.trim.clean.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}_2.trim.clean.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}.hisat2.bam'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}.hisat2.bam.bai'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}.hisat2.bam.picard.markduplicates.bam'.format(srx=srx, srr=srr))
    _delete('../output/prealn-wf/samples/{srx}/{srr}/{srr}.hisat2.bam.picard.markduplicates.bam.bai'.format(srx=srx, srr=srr))
    
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}_1.trim.clean.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}_2.trim.clean.fastq.gz'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}.fq.bam'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}.fq.bam.bai'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}.fq.first.bw'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srr}/{srr}.fq.second.bw'.format(srx=srx, srr=srr))
    _delete('../output/aln-wf/samples/{srx}/{srx}.first.bw'.format(srx=srx))
    _delete('../output/aln-wf/samples/{srx}/{srx}.second.bw'.format(srx=srx))
    
def iter_clean(df):
    cnt = 0
    for _, row in df.iterrows():
        clean(**row.to_dict())
        cnt += 1
    print('{:,} samples removed.'.format(cnt))
```
