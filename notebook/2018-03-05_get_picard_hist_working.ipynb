# Picard Hist Parser

I want to include the Picard Collect RNA Seq metrics histogram as a feature set for library startegy. I had not been parsing the hist out, so I need to add this functionality to the pre-aln store script.

```python
import os
import sys
from pathlib import Path

from IPython.display import display, HTML, Markdown
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Project level imports
sys.path.insert(0, '../lib')
from ncbi_remap.notebook import Nb
from ncbi_remap.plotting import make_figs
from ncbi_remap.parser import parse_picardCollect_hist
from ncbi_remap.snakemake import get_patterns

# Setup notebook
nbconfig = Nb.setup_notebook()

# Connect to data store
store = pd.HDFStore('../sra.h5')
```

```python
patterns = get_patterns('../prealn-wf/patterns.yaml')
```

```python
pattern = patterns['picard']['collectrnaseqmetrics']['metrics']['unstranded']
```

```python
record = store['prealn/complete'].iloc[0, :].to_dict()
```

```python
dfs = []
for idx, (srx, srr) in store['prealn/complete'].head(10).iterrows():
    fname = Path('../prealn-wf/', pattern.format(srx=srx, srr=srr))
    df = parse_picardCollect_hist(fname)
    df = df.assign(srx=srx, srr=srr)
    df.set_index(['srx', 'srr'], inplace=True)
    dfs.append(df)
```

```python
cb = pd.concat(dfs)
```
