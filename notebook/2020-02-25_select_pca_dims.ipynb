```python
import numpy as np
import pandas as pd
from joblib import load
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
reducer = load(open("../output/agg-rnaseq-wf/pca_reducer.pkl", "rb"))
```

```python
cumsum_ = np.sort(reducer.explained_variance_ratio_)[::-1].cumsum()
gte99pct = np.sum(cumsum_ < .99)

fig, ax = plt.subplots()
ax.plot(cumsum_)
ax.axvline(gte99pct, ls="--", color="k")
ax.text(gte99pct + 1000, .8, f"99% variance cutoff ({gte99pct:,})")
```
