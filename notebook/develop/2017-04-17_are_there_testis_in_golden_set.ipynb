```python
goldRes = list(biometa.aggregate([
    {'$unwind': '$experiments'},
    {'$unwind': '$experiments.runs'},
    {'$unwind': '$magic'},
    {
        '$match': {
            'magic': {'$exists': 1},
            'magic.name': 'sample summary'
        }
    },
    {
        '$project': {
            'srr': '$experiments.runs',
            'srx': '$experiments.srx',
            'magic': '$magic.value'
        }
    },
    {
        '$match': {
            'magic': {'$regex': '.*test.*'}
        }
    }
]))
```
